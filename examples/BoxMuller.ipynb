# Box-Muller Method
Rather than generating a single number, we sample a pair of random variables from 2-dimensional standard normal distribution. We start by equating equation in cartesian and polar coordinates with a 2D uniform distribution.
$$
  \int du\,dw\,\underbrace{p(u,w)}_\text{uniform}
  = \int dx\,dy\,\underbrace{p(x,y)}_\text{Gauss}
  = \int r\,dr\,d\phi \underbrace{p(r,\phi)}_\text{Gauss}
$$

$$
  \int du\,dw = \int dx\,dy \frac{1}{2\pi}e^{-\left( x^2 + y^2 \right)/2}
  = \int r\,dr\,d\phi \frac{1}{2\pi}e^{-r^2/2}
$$

$$
  du = \frac{d\phi}{2\pi}
  \qquad
  dw = r\,dr\,e^{-r^2/2}
$$

We find the coordinate transformations
$$
  \phi = 2 \pi u  
  \qquad
  r = \sqrt{-2 \ln (1-w)} = \sqrt{-2 \ln v}
$$

$$
  x = r \cos \phi
  \qquad
  y = r \sin \phi
$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
bins = [40,40]
range = [[-3,3],[-3,3]]
```

```python
x=np.random.normal(size=100000)
y=np.random.normal(size=len(x))
plt.hist2d(x,y,bins=bins,range=range);
```

# Generating $r$ and $\phi$

```python
u = np.random.uniform(size=100000)
v = np.random.uniform(size=len(u))

phi = 2*np.pi * u
r = np.sqrt(-2*np.log(v))

x = r*np.cos(phi)
y = r*np.sin(phi)

plt.hist2d(x,y, bins=[40,40], range=[[-3,3],[-3,3]]);
```
