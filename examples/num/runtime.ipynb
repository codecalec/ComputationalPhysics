```python
%matplotlib notebook
%matplotlib notebook
import numpy as np
from timeit import timeit
```

```python
nn = [1, 2, 3, 4, 5, 8, 10, 100, 1000, 5000, 10000]
tt = []

# Generate `n` random numbers and sort them.
# Measure the avg. execution time of the sorting.
for n in nn:
    niter = 100
    t = timeit( stmt = "x.sort()", 
                setup = f"import numpy ; x = numpy.random.random({n})", 
                number = niter) / niter
    tt.append(t)
```

```python
plt.figure("Run time")
plt.plot(nn,tt,"+")
plt.show()
```

```python
plt.figure("Run time (log)")
plt.loglog(nn,tt,"+")
plt.show()
```
